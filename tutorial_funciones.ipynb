### **Cómo definir una función, pasar argumentos y recopilar resultados**

```python
# Función con argumento

def func(a):
  print('I take input a :',a)

# Pasar el argumento mientras llamas a la función

a = 10
func(a)
```

### **Llamar a funciones dentro de una función y valores predeterminados de argumentos**

```python
# Supongamos que definimos una función que devuelve el valor de entrada al cuadrado
def square(a):
  return a**2

# Queremos escribir una función que calcule el valor de la hipotenusa
# Podemos llamar a la función anterior dentro de nuestra nueva función

def hypotenuse(a,b):
  hypo = (square(a) + square(b))**0.5
  return hypo

# Llamar a la función hipotenusa
a = 6
b = 8

c = hypotenuse(a,b)
```

```python
# Establecer el valor predeterminado de los argumentos

def hypotenuse(a,b=8):
  hypo = (square(a) + square(b))**0.5
  return hypo

# Ahora, al llamar a la función hipotenusa, puede pasar uno o dos argumentos
# Cuando solo se le da un argumento, asumirá el valor predeterminado para el otro

a = 6
c = hypotenuse(a)
print(c)
# Cuando se dan dos valores, no tomará el valor predeterminado para b
# Más bien, tomará el valor proporcionado por el usuario.

a = 9
b = 12
c = hypotenuse(a,b)
print(c)
```

### **Tipo de argumento, argumentos abstractos y argumentos de palabra clave.**

Puede especificar el tipo de argumento esperado, pero no se enforzará. Para hacer cumplir y enforzar, agregue un bloque que capture el error de tipo

```python
# Mencione el tipo de argumento y el tipo de retorno esperado
def arg_func(a: int) -> int:
    x = a + 10
    return x

# Pasar un float no arroja error
var1 = 6.5
var2 = arg_func(var1)
print (var2)
```

Argumentos arbitrarios, *args

```python
# Se utiliza cuando no se conoce el número de entradas
def my_function(*kids):
  print("The youngest child is " + kids[-1])

my_function("Emil", "Tobias", "Linus")

# Otro ejemplo
def average(*numbers):
  return sum(numbers)/len(numbers)

print('average for 1,2,3: ',average(1,2,3),' average for 1,10,33,43,2: ',average(1,10,33,43,2))
```

Argumentos de palabras clave arbitrarios, **kwargs

```python
# Se utiliza cuando se desconoce el número de entradas pero se espera una palabra clave específica

def my_function(**kid):
  print("His last name is " + kid["lname"])

my_function(fname = "Tobias", lname = "Refsnes")
```
